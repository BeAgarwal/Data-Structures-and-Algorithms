### Lets start visualizing the Big O Notation
We use two libraries i.e. numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

It shows the plot of a function $7n^2+6n+5$ in the range $1 \le n \le 100$. 

```python
n = np.linspace(1,100)
plt.plot(n, 7*n*n + 6*n + 5)
plt.show()
```

Now, let us add a function $20n$ to the above example to visualize that grows slower than $7n^2+6n+5$ 

```python
n = np.linspace(1, 100)
plt.plot(n, 7*n*n + 6*n +5, label="7n^2+6n+5")
plt.plot(n, 20*n, label="20n")
plt.legend(loc='upper left')
plt.show()
```

Smaller terms can be omitted.

```python
n = np.linspace(1, 100)
plt.plot(n, 7*n*n + 6*n +5, label="7n^2+6n+5")
plt.plot(n, 7*n*n, label="7n^2")
plt.legend(loc='upper left')
plt.show()
```

Lets decrease the interval from $[1,100]$ to $[1,5]$.

```python
n = np.linspace(1, 5)
plt.plot(n, 7*n*n + 6*n +5, label="7n^2+6n+5")
plt.plot(n, 7*n*n, label="7n^2")
plt.legend(loc='upper left')
plt.show()
```

As expected, $7n^2+6n+5$ is always larger than $7n^2$. 
The contribution of $6n+5$ becomes more and more negligible.

Another way of justifying this, is to plot the function $\frac{7n^2+6n+5}{7n^2}$.

```python
n = np.linspace(1,100)
plt.plot(n, (7*n*n + 6*n + 5)/(7*n*n))
plt.show()
```

As n grows, the fraction approaches $1.0$

### Multiplicative constants can be omitted.
$7n^2+6n+5 = O(n^2)$, i.e. $7n^2+6n+5$ grows no faster than $n^2$.

```python
n = np.linspace(1,100)
plt.plot(n, (7*n*n + 6*n +5)/(n*n))
plt.show()
```

Here, their fraction is always at most $18$ and approaches $7.0$ 
In other words, $7n^2+6n+5 \ge 18n^2  \space \space \forall \space \space n \le 1$

### Out of two polynomials, the one with the larger degree grows faster.

```python
n = np.linspace(1, 5)
plt.plot(n, n, label="n")
plt.plot(n, n*n, label="n^2")
plt.plot(n, n*n*n, label="n^3")
plt.legend(loc='upper left')
plt.show()
```

Now, lets increase the range from $[1, 5]$ to $[1, 100]$.

```python
n = np.linspace(1, 100)
plt.plot(n, n, label="n")
plt.plot(n, n*n, label="n^2")
plt.plot(n, n*n*n, label="n^3")
plt.legend(loc='upper left')
plt.show()
```

### Any polynomial grows slower than any exponential.
Lets plot $n^4$ and $2^n$ in the range $[1,10]$.

```python
n = np.linspace(1, 10)
plt.plot(n, n**4, label="n^4")
plt.plot(n, 2**n, label="2^n")
plt.legend(loc='upper left')
plt.show()
```

It reveals that in range $[1,10]$ $n^4$ is always greater than $2^n$. This however does not mean that $n^4$ grows faster than $2^n$. To ensure this, lets take the range $[1,20]$.

```python
n = np.linspace(1, 20)
plt.plot(n, n**4, label="n^4")
plt.plot(n, 2**n, label="2^n")
plt.legend(loc='upper left')
plt.show()
```

### Any poly-logarithm grows faster than any polynomial
Lets visualize $\log n$ , $\sqrt{n}$ and $n$. 

```python
n = np.linspace(1, 10)
plt.plot(n, n, label="n")
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n), label="log n")
plt.legend(loc='upper left')
plt.show()
```

Now, make it as $(\log{n})^3$ versus $\sqrt{n}$.

```python
n = np.linspace(1, 100)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) **3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

It seems that $(\log{n})^3$ grows faster than $\sqrt{n}$.

Lets increase the range from $[1,100]$ to say, $[1, 1000000]$. 

```python
n = np.linspace(1, 1000000)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) **3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

The logarithmic function is still above the polynomial one! 

To finally ensure that $\sqrt{n}$ outperforms $(\log{n})^3$ eventually. Lets increase the range from $10^6$ to say, $10^8$. 

```python
n = np.linspace(1, 10**8)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) **3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

Also, Lets consider as even large interval to make sure that these two functions doesn't switch back.

```python
n = np.linspace(1, 10**15)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) **3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```
